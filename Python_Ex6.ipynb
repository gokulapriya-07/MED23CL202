```python
 import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#sample data
data={
    "Student":["Aarav","Divya","Kabir","Riya","Arjun"],
    "Fee Paid":[48000,43200,45600,38400,48000]
}

df=pd.DataFrame(data)

#simple bar chart
sns.barplot(x="Student",y="Fee Paid",data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

Fees=[48000,43200,45600,38400,48000]


density=gaussian_kde(Fees)
xs=np.linspace(min(Fees),max(Fees),200)
ys=density(xs)

plt.plot(xs,density(xs))
plt.show()
```
