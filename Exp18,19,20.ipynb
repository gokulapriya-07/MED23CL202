```python
 #TYPES
 # 1.zero exception  error

try:
  print(10/0)
except ZeroDivisionError:
    print("Zerodivision error : You cannot divide by zero.\n")

# 2.Valueerror

try:
  num=int("abc")
except ValueError:
  print("ValueError: Cannot convert string to integer.\n")

# 3.Typeerror

try:
  result="2"+2
except TypeError:
  print("Typeerror: you tried to add string and integer.\n")

# 4.IndexError

try:
  lst=[1,2,3]
  print(lst[5])
except IndexError:
  print("IndexError: List index out of range.\n")

# 5.KeyError

try:
  my_dict={'name':'Anu'}
  print(my_dict['age'])
except KeyError:
  print("KeyError: Key not found in dictionary.\n")

#6.FilenotfoundError

try:
  open("missing.txt")
except FileNotFoundError:
  print("FileNotFoundError: File not found.\n")


# 7.NameError

try:
  print(undefined_variable)
except NameError:
  print("NameError: Variable is not defined.\n")
```
