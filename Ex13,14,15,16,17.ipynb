```python

with open("/content/sampledata.txt","r")as f:
  print(f.tell())
  data=f.read(6)
  print(data)
  print(f.tell())
```

```python
with open("/content/sampledata.txt","r")as f:
  print(f.seek(10))
  data=f.read(6)
  print(data)
  print(f.seek(0))
```

```python
#to check file mode
with open("/content/sampledata.txt","w")as f:
  print(f.seekable())
  print(f.writable())
  print(f.readable())
```

```python
#re-naming the file
import os
with open("/content/oldfile.txt", "w") as f:
    f.write("This is the old file content.")

os.rename("/content/oldfile.txt","newfile.txt")
print("File renamed successfully")

#deleting a file
os.remove("newfile.txt")
print("File deleted successfully")
```
