```python
import matplotlib.pyplot as plt
students=["Ramya","Thamizh","Jaya","Seenu","Priya"]
fees=[40000,45000,32000,43000,50000]
plt.barh(students,fees,color="skyblue")
plt.title("Fees paid by students")
plt.xlabel("Fees Paid")
plt.ylabel("Students")
plt.show()
```

PIE CHART

```python
import matplotlib.pyplot as plt
students=["Ramya","Thamizh","Jaya","Seenu","Priya"]
fees=[40000,45000,32000,43000,50000]
colors=["skyblue","lightgreen","gold","coral","lightcoral"]
plt.pie(fees,labels=students,colors=colors,autopct='%2.2f%%',startangle=80)
plt.title("Fees paid by students")
plt.show()
```

Interactive Visualisation
