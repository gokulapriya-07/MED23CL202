Dictionary CRUD operations

CREATE ( add new key value pair)

```python
student={}
student["name"]="Alice"
student["age"]=19
print(student)
```

READ

```python
print(student["name"])   #read the key
print(student.get("age"))
```

```python
print(student.get("grade"))
```

```python
print(student.get("grade","not avaliable"))
```

UPDATE

```python
student={}
student["name"]="Bob"       #update the key name , its value is bob
student["age"]=19
student["grade"]='A'
print(student)
```

DELETE

```python
del student["grade"]      # it will completely removes the key and value
print(student)
```

```python
#another way to delete
remove=student.pop("age")    #pop will delete only its key , the value will present but cannot be accessed
print(student)
print(remove)
```

NESTED DICTIONARY USING STUDENT DETAILS

```python
student={
    "Alice":{"Math":53,"Science":90,"English":80},
    "Bob":{"Math":87,"Science":50,"English":54},
    "Charlie":{"Math":89,"Science":78,"English":99},
    "Alex":{"Math":49,"Science":98,"English":49},
    "Selena":{"Math":72,"Science":38,"English":49}
}
print(student)
```

for data manipulation using dictioanry convert the dictionary to dataframe

```python

import pandas as pd #pandas are library for data manipulation
data=pd.DataFrame(student)  #pass the dictionary to DataFrame
data
```

```python
data=pd.DataFrame(student).T    #transpose to get the student as row
data
```

```python
data["Average"]=data.mean(axis=1)   #axis=1 denotes the row wise access to the dataframe
data
```

```python
data["Math"]=data.mean(axis=0)   #axis=0 denotes column wise access to the dataframe
data    # the result is not a number(NaN) because there is no variable to store the average of
```

Data Visualizations

```python
import matplotlib.pyplot as plt  #bar chart for student average performance
plt.figure(figsize=(8,5))
data["Average"].plot(kind="bar",color="skyblue",edgecolor="black")
plt.title("Average Marks of Student")
plt.xlabel("Student")
plt.ylabel("Average Score")
plt.show()
```

```python

plt.figure(figsize=(8,5))  #
for subject in["Math","Science","English"]:
  plt.plot(data.index,data[subject],marker='o',label=subject)
plt.title("Subjectwise Performance")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python

plt.figure(figsize=(8,5))  #
data[["Math","Science","English"]].plot(kind="box")
plt.title("Marks Dustribution by Students")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```
